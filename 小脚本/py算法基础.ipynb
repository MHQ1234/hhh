```python
class Queue:
    
    def __init__(self):
        self.items = []
    
    def isEmpty(self):
        return self.items == []
    
    def enqueue(self, item):
        self.items.insert(0, item)
        
    def dequeue(self):
        return self.items.pop()
    
    def size(self):
        return len(self.items)
    
    def values(self):
        return self.items
    
q = Queue()
q.enqueue(3)
q.size()
q.isEmpty()
q.dequeue()
q.isEmpty()
q.values()
```

队列应用——热土豆问题

```python
def hotpotato(namelist, num):
    
    simqueue = Queue()#实例化
    
    for name in namelist:
        simqueue.enqueue(name)#将人名依次加入到队列中
        
    while simqueue.size() > 1:
        for i in range(num):
            simqueue.enqueue(simqueue.dequeue())#达到固定次数之前，传到的每一个人从列首移动到列尾，形成闭环
        simqueue.dequeue()#一轮游戏结束，不幸的玩家出列，剩下的人继续重复，直到剩下一个人为止
        
    return simqueue.dequeue()#删除并返回队列中的唯一幸存玩家

namelist = ['1','2','3','4','5','6']
hotpotato(namelist, 6)
```

```python
class Deque:
    
    def __init__(self):
        self.items = []
    
    def isEmpty(self):
        return self.items == []
    
    def size(self):
        return len(self.items)
    
    def values(self):
        return self.items
    
    def addRear(self, item):
        self.items.insert(0, item)
    
    def addFront(self, item):
        self.items.append(item)
        
    def removeRear(self):
        return self.items.pop(0)
    
    def removeFront(self):
        return self.items.pop()
D = Deque()
D.isEmpty()
D.size()
D.addRear(1)
D.addFront(2)
D.values()
D.removeFront()
D.values()
D.removeRear()
D.values()
```

双端队列应用——回文词判定

```python
def palchecker(aString):
    chardeque = Deque()
    
    for ch in aString:
        chardeque.addRear(ch) #将字符串分成字符并赋值到列表
    
    flag = True
    while chardeque.size() > 1 and flag:
        if chardeque.removeRear() != chardeque.removeFront():
            flag = False #判断字符串的收尾是否相同，若相同则取出收尾，然后重复过程，若不同，则停止迭代
    return flag #返回字符串是否为回文词

print(palchecker('abndnba'))
print(palchecker('天连水尾水连天'))
print(palchecker('lalala'))
        
```

```python
def divideBy2(decNumber, k):
    remstack = []
    while decNumber > 0:
        rem = decNumber % k # 求余数
        remstack.append(str(rem)) # 将余数添加到栈内，形成二进制数值末端的值
        decNumber = decNumber // k # 求除数，并将其代替原来的数值
    return int("".join(remstack[::-1]))
divideBy2(35, 2)
```

冒泡排序

```python
# 最优复杂度：O(n)
# 最差复杂度：O(n^2)
# 平均复杂度: O(n^2)
# 逆序数越少，算法越快

def bubble_sort(alist):
    n = len(alist)
    for i in range(n-1):
        count = 0 # 计数器，用来终止无意义的循环，减少时间复杂度
        for j in range(n-1-i):
            if alist[j] > alist[j+1]:
                alist[j], alist[j+1] = alist[j+1], alist[j]
                count +=1 # 如果有交换发生，记录次数
        if count == 0:# 如果没有交换发生，排序已完成
            return

alist = list(range(10000, -1, -2)) + list(range(1, 10000, 2))
# alist = list(range(100000, -1, -1))
import time
time_1 = time.time()
bubble_sort(alist)
# print(slist)
time_2 = time.time()
print('time:'+ str(time_2-time_1))
alist
```

选择排序

```python
# 最优复杂度：O(n^2)
# 最差复杂度：O(n^2)
# 平均复杂度: O(n^2)
# 对比冒泡和插入法，逆序数越多，选择算法效率越高，逆序数越少，效率越低

def select_sort(alist):
    n = len(alist)
    for i in range(n-1):
        min_index  = i
        for j in range(i, n):
            if alist[j] < alist[min_index]: min_index = j
        alist[i], alist[min_index] = alist[min_index], alist[i]

# alist = list(range(100000, -1, -2)) + list(range(1, 100000, 2))
alist = list(range(1000, -1, -1))
import time
time_1 = time.time()
select_sort(alist)
time_2 = time.time()
print('time:'+ str(time_2-time_1))
# alist
```

插入排序

```python
# 最优复杂度：O(n)
# 最差复杂度：O(n^2)
# 平均复杂度: O(n^2)
# 逆序数越少，算法越快

def insert_sort(alist):
    n = len(alist)
    for i in range(n-1):
        for j in range(i+1, 0,-1):
            if alist[j] < alist[j-1]:
                alist[j], alist[j-1] = alist[j-1], alist[j]
            else:
                break

alist = list(range(100000, -1, -2)) + list(range(1, 100000, 2))
# alist = list(range(100000, -1, -1))
import time
time_1 = time.time()
insert_sort(alist)
time_2 = time.time()
print('time:'+ str(time_2-time_1))
alist
```

希尔排序

```python
# 最优复杂度：O(n^1.3)
# 最差复杂度：O(n^2)
# 平均复杂度：O(nlog(n)~n^2)(不稳定)

def shell_sort(alist):
    n = len(alist)
    gap = n//2
    while gap > 0:
        for i in range(n-gap):
            for j in range(i+gap, 0, -gap):
                if alist[j] < alist[j-gap]:
                    alist[j], alist[j-gap] = alist[j-gap], alist[j]
                else:
                    break
        gap = gap//2

alist = list(range(100000, -1, -2)) + list(range(1, 100000, 2))
# alist = list(range(100000, -1, -1))
import time
time_1 = time.time()
shell_sort(alist)
time_2 = time.time()
print('time:'+ str(time_2-time_1))
# alist
```

快速排序

```python
# 最优复杂度：O(nlog(n))
# 最差复杂度：O(n^2)
# 平均复杂度：O(nlog(n))

# 设置最大递归次数
import sys
sys.setrecursionlimit(100000) # 设置限制数100000

def quick_sort(alist, first, last):
    if first >= last: return
    mid_value = alist[first]
    low, high = first, last
    while low < high:
        while low < high and alist[high] >= mid_value: high -= 1 # 游标high左移
        alist[low] = alist[high]
        while low < high and alist[low] <= mid_value: low += 1 # 游标low右移
        alist[high] = alist[low]
    alist[low] = mid_value
#     摆正位置的元素不参与下次迭代！！！
    quick_sort(alist, first, low-1) # 对low左边的列表执行快速排序
    quick_sort(alist, low+1, last) # 对low右边的列表排序
    
alist = list(range(100000, -1, -2)) + list(range(1, 100000, 2))
# alist = list(range(100000, -1, -1))
import time
time_1 = time.time()
quick_sort(alist, 0, len(alist)-1)#递归函数，没有返回值
time_2 = time.time()
print('time:'+ str(time_2-time_1))
# alist
```

```python
# 最优复杂度：O(nlog(n))
# 最差复杂度：O(nlog(n))
# 平均复杂度：O(nlog(n))
# 需要辅助空间（申请其他变量用来存储中间结果）

import time
def merge_sort(alist):
    if len(alist) <= 1: return alist
    mid = len(alist)//2
    return merge(merge_sort(alist[:mid]), merge_sort(alist[mid:]))
def merge(left, right):
    l, r = 0, 0
    result = []
    while l < len(left) and r < len(right):
        if left[l] < right[r]:
            result.append(left[l])
            l += 1
        else:
            result.append(right[r])
            r += 1
    result += left[l:]
    result += right[r:]
    return result

alist = list(range(1000000, -1, -2)) + list(range(1, 1000000, 2))
time_1 = time.time()
# alist = list(range(100000, -1, -1
alist = merge_sort(alist)#递归函数，没有返回值
time_2 = time.time()
print('time:'+ str(time_2-time_1))
# alist
```

堆排序

```python
# 堆排序是一种完全二叉树，分为大根堆和小根堆，大根堆的每个父节点值逗比子节点大，小根堆相反
# 将二叉树的值转为列表存储的方式：1.根节点在第一位（索引是0）2.父节点的索引是i，则左子节点的索引为2i+1，右子节点的索引为2i+2
# 算法复杂度：
# 最优：O(nlog(n))
# 最差：O(n^2)
# 平均：O(nlog(n))

# 堆排序的向下调整：
def sift(li, low, high):
    '''
    li:列表
    Low:堆的根节点位置（第一个元素）
    high:堆的最后一个元素的位置
    '''
    i = low # 最开始指向根节点
    j = 2 * i + 1 # j最开始指向i节点的左孩子
    tmp = li[i] # 存储根节点元素
    while j <= high: # 只要左子节点不出界
        if j+1 <= high and li[j+1] > li[j]: # 如果右子节点也不出界并且右子节的元素值点大于左子节点的
            j = j + 1 # 则将j指向更大的右子节点
        if li[j] > tmp: # 如果子节点的值比父节点大
            li[i] = li[j] # 那么将父节点的值替换为子节点的值
            i = j # 同时游标i下移到子节点
            j = 2 * i + 1 # j也下移到相应的新的子节点
        else: # 如果子节点不大于父节点，停止调整
            break
    li[i] = tmp # 无论是出界还是停止调整，都需要将最后一个节点赋值
    
def heap_sort(li):
    n = len(li)
    for i in range((n-2)//2, -1, -1):
        sift(li, i, n-1)
    for j in range(n-1, -1, -1):
        li[j], li[0] = li[0], li[j]
        sift(li, 0, j-1)
        
alist = list(range(100000, -1, -2)) + list(range(1, 100000, 2))
# alist = list(range(100000, -1, -1))
import time
time_1 = time.time()
heap_sort(alist)#递归函数，没有返回值
time_2 = time.time()
print('time:'+ str(time_2-time_1))
alist
```

计数排序

```python
# 时间复杂度：O(n)
# 限制：1.只能对整数使用 2.当列表内元素值极大时，需要申请很大的额外辅助空间

def count_sort(alist):
    min_value = min(alist)
    max_value = max(alist) #求极大极小值可以放在外面，在里面会占用2n的复杂度
    count_list = [0] * (max_value - min_value + 1)
    for val in alist:
        count_list[val - min_value] += 1
    alist.clear()
    for i in range(min_value, max_value + 1):
        for j in range(count_list[i - min_value]):
            alist.append(i)

alist = list(range(100000, -1, -2)) + list(range(1, 100000, 2))
# alist = list(range(100000, -1, -1))
import time
time_1 = time.time()
count_sort(alist)#递归函数，没有返回值
time_2 = time.time()
print('time:'+ str(time_2-time_1))
alist
```

桶排序

```python
# 时间复杂度：O(kn)
# 限制：只能处理非负整数

def bin_sort(alist):
    import numpy as np
    max_value = max(alist)
    n = int(np.sqrt(max_value))
    bin_num = max_value//n
    bin_list = [[] for i in range(n)]
    for val in alist:
        v = min(val//bin_num, n-1)
        bin_list[v].append(val)
        for j in range(len(bin_list[v])-1, 0, -1):
            if bin_list[v][j] < bin_list[v][j-1]:
                bin_list[v][j], bin_list[v][j-1] = bin_list[v][j-1], bin_list[v][j]
    alist.clear()
    for b in bin_list:
        alist.extend(b)

# alist = list(range(100000, -1, -2)) + list(range(1, 100000, 2))
alist = list(range(100000, -1, -1))
import time
time_1 = time.time()
bin_sort(alist)#递归函数，没有返回值
time_2 = time.time()
print('time:'+ str(time_2-time_1))
alist
    
```

基数排序

```python
# 时间复杂度：O(kn)，k是最大元素值的位数
# 限制：只能用于非负整数的排序以及字符串排序，但可以自行改良

def cardinal_sort(alist):
    max_value = max(alist)
    it = 0
    while 10 ** it <= max_value:
        bin = [[] for i in range(10)]
        for val in alist:
            digit = val//(10 ** it) % 10
            bin[digit].append(val)
        alist.clear()
        for val in bin:
            alist.extend(val)
        it += 1
        
alist = list(range(100000, -1, -2)) + list(range(1, 100000, 2))
# alist = list(range(100000, -1, -1))
import time
time_1 = time.time()
cardinal_sort(alist)#递归函数，没有返回值
time_2 = time.time()
print('time:'+ str(time_2-time_1))
alist
```

```python
# lowb三人组：冒泡，选择，插入
# 速度对比：平均来讲复杂度都是O(n^2)，差距不大，最好的情况下冒泡和插入比较快 

# nb三人组：快速、归并、堆排序
# 速度平均来讲：快速>归并>堆
# 速度不同的原因：虽然复杂度量级相同，但是系数不同
# 缺点方面：
# 1.快排：极端情况下效率低（完全逆序，此时复杂度是n^2）
# 2.归并：需要额外的内存开销
# 3.堆：相对来讲速度最慢

# 希尔排序：介于lowb和Nb之间，是改良版的插入排序

# 以上排序是最常使用的，都是基于数值间比较的原理实现的

# 特殊三人组：计数、分桶、基数排序（不是基于比较原理实现的排序）
# 共同特点是只能处理整数排序，使用场景的比较局限，但是在特定场景可以实现线性复杂度
# 虽然在特定场景下这三种排序可以有线性复杂度O(kn)，但并不代表它们一定比nb三人组更快，因为k是不确定的，
# 在n有限的情况下，k足够大时会出现nb三人组更快的情况
```

二叉搜索树

```python
# 定义：在一个树种，对于任何一个节点，它左子树的所有元素值都比它小，右子树的所有元素值都比它大
# 树内一般没有重复节点值，所以不考虑等于的情况

class BST:
    def __init__(self,li = None):
        self.root = None
        if li:
            for val in li:
                self.insert(val)
                
    class BiTreeNode:
        def __init__(self, data):
            self.data = data
            self.lchild = None
            self.rchild = None
            self.parent = None
            
    def insert(self, val): #插入
        p = self.root
        if not p:
            self.root = BiTreeNode(val)
            return
        while True:
            if val < p.data:
                if not p.lchild:
                    p.lchild = BiTreeNode(val)
                    p.lchild.parent = p
                    return
                else:
                    p = p.lchild
            elif val > p.data:
                if not p.rchild:
                    p.rchild = BiTreeNode(val)
                    p.rchild.parent = p
                    return
                else:
                    p = p.rchild
            else:
                return
    
    def query(self, val): #查询
        p = self.root
        while p:
            if p.data == val:
                return p
            elif p.data > val:
                p = p.lchild
            else:
                p = p.rchild
        else:
            return None
            
    def pre_order(self, root): #前序遍历
        if root:
            print(root.data, end = ',')
            self.pre_order(root.lchild)
            self.pre_order(root.rchild)

    def in_order(self, root): #中序遍历
        if root:
            self.in_order(root.lchild)
            print(root.data, end = ',')
            self.in_order(root.rchild)

    def beh_order(self, root): #后续遍历
        if root:
            self.beh_order(root.lchild)
            self.beh_order(root.rchild)
            print(root.data, end = ',')

    def level_order(self, root): #层次遍历
        queue = []
        queue.append(root)
        while len(queue) > 0:
            Node = queue.pop(0)
            print(Node.data, end = ',')
            if Node.lchild:
                queue.append(Node.lchild)
            if Node.rchild:
                queue.append(Node.rchild)
    
    

import random
li = list(range(10))
# random.shuffle(li)
print('li:', li)
tree = BST(li)
# tree.pre_order(tree.root)
# tree.in_order(tree.root)
# tree.beh_order(tree.root)
tree.level_order(tree.root)
```

```python
# 找零问题：给定额度的零钱，用最小数量的货币表示出来
# 思想：从大的面值开始找钱，面值越大所需数量越少
def change(value):
    t = [100, 50, 20, 10, 5, 1]
    v_num = [0 for i in range(len(t))]
    c = 0
    while c < len(t) and value != 0:
        num = value//t[c]
        v_num[c] = num
        value -= num * t[c]
        c += 1
    print('the min num is: ', sum(v_num))
    return list(zip(t, v_num))
change(1)
```

```python
# 背包问题：有n件价值为各自为vi，重量各自为wi的宝贝可装进背包，但背包重量限制W（假设体积不限），怎么装可实现价值最大
# 背包问题分为两种，0-1问题和比例问题，分别对应物品是否可以拆分的两种情况
# 思想:计算每件宝贝的单位重量下的价值，从大到小排序，越大的越优先考虑

def bag(v, w, W): # v是价值列表，w是重量列表，W是背包重量限制
    n = len(v)
    u = []
    for i in range(n):
        u.append(v[i]/w[i])
    c = list(zip(u, v, w))
    c = sorted(c, reverse=True)
    u1, v1, w1 = zip(*c)
    nums = [0 for i in range(n)]
    total_w = 0
    count = 0
    while count < n:
        if total_w + w1[count] <= W:
            total_w += w1[count]
            nums[count] = 1
        else:
#             nums[count] = round((W - total_w)/w1[count],2) #如果是0-1背包问题，不需要这行代码，如果是比例背包问题则需要
            break
        count += 1
    return list(zip(v1, w1, nums))

v = [1,2,3,4,5,7,3,6,8,4,2,6,89,0]
w = [4,3,5,2,6,2,6,7,8,94,3,54,3,2]
W = 10
bag(v, w, W)
```

```python
# 拼接数字问题：若干数字，都是非负整数，大小长短不一，求拼接后所得最大整数
# 问题关键：对于128和1286等这种一个数字是另一个的子串的问题，如何解决
# 思想：思想就是排序，不过判断条件是比较相邻两个数字，互相收尾拼接后看哪个更大
# 算法：类似于插入排序
def pinjie(alist):
    slist = list(map(str, alist))
    n = len(alist)
    for i in range(n-1):
        for j in range(i+1, 0, -1):
            if (slist[j]+slist[j-1]) > (slist[j-1]+slist[j]):
                slist[j], slist[j-1] = slist[j-1], slist[j]
            else:
                break
    return list(map(int, slist))
pinjie([728, 7286, 1286, 128])
```

```python
# 活动选择问题：在有限的时间内（0-N表示）安排最多的活动，每个活动有其各自开始结束时间(a-b)，活动不能重复
# 思想：最先结束的那个活动必然出现在最优解中，确定他后，在剩余的活动中重复此过程（注意不能重复即可）

# 方法1：先从所有活动中选出最小的结束时间存在list，然后找出所有开始时间>=此结束时间的活动重新存成一个list，再从这些活动中找到最小的结束时间，重复
def huodong(alist): #输入各个活动的起始时间和截止时间
    import copy
    hd = []
    bx = copy.deepcopy(alist)
    while True:
        min_index = 0
        for i in range(len(bx)):
            if bx[i][1] < bx[min_index][1]:
                min_index = i
        hd.append(bx[min_index])
        bx.clear()
        for i in range(len(alist)):
            if alist[i][0] >= hd[-1][1]:
                bx.append(alist[i])
        if len(bx) == 0:
            break
    return hd

# 方法2：先对所有活动按照结束时间从小到大排序，然后选出最小加入list1，再对接下来的活动按顺序判断是否其开始时间>=list1中最后一个活动的结束时间，如果是则加入list1,重复
def hd(alist):
    alist.sort(key = lambda x:x[1])
    res = [alist[0]]
    for i in range(1, len(alist)):
        if alist[i][0] >= res[-1][1]:
            res.append(alist[i])
    return res

alist = [(1,4),(3,5),(0,6),(5,7),(3,9),(5,9),(6,10),(8,11),(8,12),(2,14),(12,16)]
huodong(alist)
hd(alist)
```

```python
# 钢条切割问题

p = [0, 1, 5, 8, 9, 10, 17, 17, 20, 21, 23, 24, 26, 27, 27, 28, 30, 33, 36, 39, 40]

# 第一版代码：采用最直观的递推式进行递归运算，会有子结构重复运算的问题，复杂度是指数级
def cut_rod_recurision_1(p, n):
    if n == 0:
        return 0
    else:
        res = p[n]
    for i in range(1, n):
        res = max(res, cut_rod_recurision_1(p, i) + cut_rod_recurision_1(p, n-i))
    return res

# 第二版代码：相比于第一版，在递推式上做了改进，子结构重复问题减轻，但仍然有，复杂度仍是指数级
def cut_rod_recurision_2(p, n):
    if n == 0:
        return 0
    else:
        res = 0
        for i in range(1, n+1):
            res = max(res, p[i] + cut_rod_recurision_2(p, n-i))
        return res

# 第三版代码：采用第二版递推式，但不做递归，自下而上一步步得到结果后保存用于更大级别使用，消除重复子结构问题，复杂度是n^2级别
def cut_rod_dp(p, n):
    r = [0]
    for i in range(1, n+1):
        res = 0
        for j in range(1, i+1):
            res = max(res, p[j] + r[i - j])
        r.append(res)
    return r[n]

# 第四版代码：在第三版代码基础上加了个输出切割路径
def cut_rod_extend(p, n):
    r = [0]
    s = [0]
    for i in range(1, n+1):
        res_r = 0
        res_s = 0
        for j in range(1, i+1):
            if p[j] + r[i-j] > res_r:
                res_r = p[j] + r[i - j]
                res_s = j
        r.append(res_r)
        s.append(res_s)
    return r[n], s

def cut_rod_solution(p, n):
    r, s = cut_rod_extend(p, n)
    ans = []
    while n > 0:
        ans.append(s[n])
        n -= s[n]
    return r, ans

import time
time_1 = time.time()
# print(cut_rod_recurision_1(p, 15))
# print(cut_rod_recurision_2(p, 15))
# print(cut_rod_dp(p, 15))
# print(cut_rod_solution(p, 15))
time_2 = time.time()
print('time:'+ str(time_2-time_1))
```
