1 单变量分析函数编写

1.1 特征风险指标构建--基于等频分箱

```python
#函数说明：
#此函数通过输入数据集和某个（一次只能一个）特征（要数据，不能只输入名字），以及分箱数量和其他相关参数，对数据进行分箱操作，并建立各种指标，最终以df类型数据框输出，并输出相关图像，帮助检查fpd排序性
#此函数内部会自己处理缺失值，不需要提前填充，如果要提前填充也可以，但是只能填充为字符型'nan'，否则会报错，后面的三个函数也是如此。（由于这些缺失值一般不填充为其他值，就以空值单独成一组分析，所以需要填充为字符型）
def qcut_bin(df, var, y, tp1 = 0, n = 1, tp2 = 0): 
    #df是数据集，var是单个特征；y代表'是否已经逾期；n是箱数；tp1:1为定类，0为定量，2为不分箱的定序，默认0；tp2是nan值在列联表的位置，1表示在底部，0表示在顶部
    
    #调用需要的包
    import numpy as np
    import pandas as pd
    import scipy
    import matplotlib.pyplot as plt
    
    #如特征有缺失值，填补为‘nan’
    if var.isnull().sum() > 0:
        var = var.fillna('nan') #将null填充为nan
        var[var == ''] = 'nan'
        
    #定义ks函数，以备下面使用
    def ks_2(var, y):
        from scipy.stats import ks_2samp
        var = var[var != 'nan']
        y = y[var.index]
        ks = ks_2samp(var[y == 1], var[y != 1]).statistic
        return ks
    
    #定义函数：将小数转化为百分数（字符型）
    f = lambda x: format(x, '.2%') 
    
    #进入正式流程
    if tp1 == 0: #如果var是连续变量
        ks = ks_2(var, y) #引用ks_2函数计算ks值
        if (var == 'nan').sum() != 0 : #如果变量里面有nan值，需要进行如下处理
            df['qcut'], updown = pd.qcut(var[var != 'nan'], q = n, retbins = True, duplicates = 'drop') #等频分箱函数，当箱子边界值不唯一时，设置duplicates = 'drop'
            t1 = pd.crosstab(df.qcut, y) #列是分箱维度，行是y2,y1二级维度的列联表
            t2 = pd.crosstab(var[var == 'nan'], y[var == 'nan']) #为nan组的数据单独创建列联表
            t2.index.name = 'qcut' 
            t3 = pd.concat([t2, t1], axis = 0) #将两个列联表合并，得到完整的列联表（注意：只能是单行列联表在上，可能是某些语法规定）
            t3.loc['nan', 1] = np.where(np.isnan(t3.loc['nan', 1]) == True, 0, t3.loc['nan', 1]) #如果恰好特征是nan的时候没有fpd7发生，列联表中nan行的fpd7=1的计数是nan,需要换成0（因为nan是单独创建列联表再拼接，如果没有fpd7，列联表就不会显示fpd7=1这个类别）
            is_satis_fpd = t3.sum(axis = 1) #计算满足逾期条件的数目
            is_fpd = t3.loc[:, 1] #计算发生逾期的次数
            risk_per_0 = is_fpd/is_satis_fpd # 计算fpd比率
            risk_per_1 = pd.concat([risk_per_0[1:], risk_per_0[0:1]], axis = 0) #将nan组的数据移动到最下面，为后面画图做准备
            risk_per_out = risk_per_0.apply(f) #将风险比率化成百分数
            
            #以下为计算IV的过程
            y_0_rate = (t3/t3.sum()).iloc[:,0] #返回每个分箱内好客户占所有好客户的比例
            y_1_rate = (t3/t3.sum()).iloc[:,1] #返回每个分箱内坏客户占所有坏客户的比例
            woe = np.log((y_0_rate+0.0001)/(y_1_rate+0.0001)) #计算每个箱内的woe值，为防止出现箱内只有一种标签使得log计算出错，将每个比例后面都加0.0001
            iv_bin = ((y_0_rate - y_1_rate) * woe)
            iv = iv_bin.sum()
            sample_rate = (t3.sum(axis = 1)/t3.sum(axis = 1).sum()).apply(f) #等频分箱每个箱体的样本比率
            
            d_fpd = pd.concat([sample_rate, is_satis_fpd, is_fpd, risk_per_out, woe, iv_bin], axis = 1)
            d_fpd.index.name = var.name + '_bin'
            d_fpd.columns = ['各箱样本比率', '满足fpd7人数', 'fpd7人数', 'fpd7', 'woe', '各箱iv']
            d_iv = pd.DataFrame(index = d_fpd.index, columns = ['总体iv']).fillna(iv)
            d_ks = pd.DataFrame(index = d_fpd.index, columns = ['ks']).fillna(ks) #创建ks字段
            d_fpd['总体iv'], d_fpd['ks'] = d_iv, d_ks
            d_fpd1 = pd.concat([d_fpd.iloc[1:, :], d_fpd.iloc[0:1, :]], axis = 0) #将nan所在行放到最下面，当nan的数据在意义上更接近最大箱体数据时候，为使风险排序单调，需调整顺序
            
            if tp2 == 0: #如果要把nan组的数据放在数据框的顶部
                result = d_fpd.reset_index()
                for i in range(1, result.shape[0]): #将分箱数据改一下，每一个箱体前加一个编号，以表明是第几个箱子，如'01(1,2]', 下同
                    result.iloc[:, 0][i] = '0' + str(i) + str(result.iloc[:, 0][i])
            else:
                result = d_fpd1.reset_index()
                for i in range(result.shape[0]-1):
                    result.iloc[:, 0][i] = '0' + str(i+1) + str(result.iloc[:, 0][i])
                    
        else: #没有nan值得连续变量进行如下处理
            df['qcut'], updown = pd.qcut(var, q = n, retbins = True, duplicates = 'drop')
            t1 = pd.crosstab(df.qcut, y)
            t1.index.name = var.name + '_bin'
            is_satis_fpd = t1.sum(axis = 1)
            is_fpd = t1.loc[:, 1]
            risk_per_2 = is_fpd/is_satis_fpd
            risk_per_out = risk_per_2.apply(f)
            y_0_rate = (t1/t1.sum()).iloc[:,0] 
            y_1_rate = (t1/t1.sum()).iloc[:,1] 
            woe = np.log((y_0_rate+0.0001)/(y_1_rate+0.0001)) 
            iv_bin = ((y_0_rate - y_1_rate) * woe)
            iv = iv_bin.sum()
            sample_rate = (t1.sum(axis = 1)/t1.sum(axis = 1).sum()).apply(f)

            d_fpd = pd.concat([sample_rate, is_satis_fpd, is_fpd, risk_per_out, woe, iv_bin], axis = 1)
            d_fpd.index.name = var.name + '_bin'
            d_fpd.columns = ['各箱样本比率', '满足fpd7人数', 'fpd7人数', 'fpd7', 'woe', '各箱iv']
            d_iv = pd.DataFrame(index = d_fpd.index, columns = ['总体iv']).fillna(iv)
            d_ks = pd.DataFrame(index = d_fpd.index, columns = ['ks']).fillna(ks) #创建ks字段
            d_fpd['总体iv'], d_fpd['ks'] = d_iv, d_ks
            result = d_fpd.reset_index()
            
            #涉及到pd.interval的类型数据，总是很奇怪，正常手段难以重置列，只能先删再加
            new_c = ['0' + str(i+1) + str(result.iloc[:, 0][i]) for i in range(result.shape[0])]
            result.pop(result.iloc[:, 0].name)
            result.insert(0, (var.name + '_bin'), new_c)
        
        #画出风险比率趋势图
        risk_per = risk_per_2 if (var == 'nan').sum() == 0 else risk_per_0 if (tp1 == 0) & (tp2 == 0) else risk_per_1
        num = range(1, risk_per.shape[0]+1)
        
        #这两行代码解决 plt 中文显示的问题
        plt.rcParams['font.sans-serif'] = ['SimHei']
        plt.rcParams['axes.unicode_minus'] = False
        plt.figure(figsize = (15,5))
        plt.barh(result.iloc[:, 0], risk_per, label = 'qcut_bin_num = ' + str(risk_per.shape[0]))  
        plt.xlabel('风险比率')
        plt.title('各箱风险比率')
        plt.show()
        print('箱体边界值(不含nan)：', list(updown))
        print('各箱风险比率：',list(risk_per.apply(f)))
        print('总体IV值：', iv)
        
    else: #如果是定类离散变量，进行如下处理
        t = pd.crosstab(var, y)
        is_satis_fpd = t.sum(axis = 1)
        is_fpd = t.loc[:, 1]
        risk_per = is_fpd/is_satis_fpd
        risk_per_out = risk_per.apply(f)
        y_0_rate = (t/t.sum()).iloc[:,0] 
        y_1_rate = (t/t.sum()).iloc[:,1] 
        woe = np.log((y_0_rate+0.0001)/(y_1_rate+0.0001)) 
        iv_bin = ((y_0_rate - y_1_rate) * woe)
        iv = iv_bin.sum()
        sample_rate = (t.sum(axis = 1)/t.sum(axis = 1).sum()).apply(f)
        
        d_fpd = pd.concat([sample_rate, is_satis_fpd, is_fpd, risk_per_out, woe, iv_bin], axis = 1)
        d_fpd.index.name = var.name
        d_fpd.columns = ['各箱样本比率', '满足fpd7人数', 'fpd7人数', 'fpd7', 'woe', '各箱iv']
        d_iv = pd.DataFrame(index = d_fpd.index, columns = ['总体iv']).fillna(iv)
        d_fpd['总体iv'] = d_iv
        
        if tp1 == 2:
            ks = ks_2(var, y) #引用ks_2函数计算ks值
            d_ks = pd.DataFrame(index = d_fpd.index, columns = ['ks']).fillna(ks)
            d_fpd['ks'] = d_ks
            
        result = d_fpd.reset_index()
        
        #画出风险比率条形图
        #这两行代码解决 plt 中文显示的问题
        plt.rcParams['font.sans-serif'] = ['SimHei']
        plt.rcParams['axes.unicode_minus'] = False
        # 横放条形图函数 barh
        plt.figure(figsize = (12,8))
        plt.barh(result.iloc[:, 0], risk_per)  
        plt.xlabel('风险比率')
        plt.title('各类别风险比率')
        plt.show()
        print('各类别风险比率：\n', list(zip(result.iloc[:, 0], risk_per_out)))
        print('总体IV值：', iv)
        
        return result 
    
    return result, updown #只要是连续特征，需要返回箱体边界值，以用来后面进行可能的woe编码
    
#函数说明，此函数建立在qcut_bin函数之上使用，意义在于设置最大箱数n，并调用qcut_bin函数输出箱数属于1-n之间的所有结果集，并输出相关图像
#不适用于定类变量！
def qcut_bin_all(df, var, y, max_n = 1, tp = 0): #df是数据集，var是目标变量；y代表'是否已经逾期'；n是最大箱数；tp是nan值在列联表的位置，1表示在底部，0表示在顶部
    #检查该特征是否有缺失值，如有则进行填补
    if var.isnull().sum() > 0:
        var = var.fillna('nan') #将null填充为nan
        var[var == ''] = 'nan'
    #获取结果  
    result_all = []
    upd = []
    for i in range(1, max_n+1):
        result = qcut_bin(df, var, y, tp1 = 0, n = i, tp2 = tp)
        result_all.append(result[0])
        upd.append(result[1])
    #画图输出IV值变化趋势
    iv_list = [] #定义IV值列表
    for i in range(len(result_all)):
        iv_list.extend(list(set(list(result_all[i]['总体iv'])))) #获取IV值列表
    #进行画图
    num = range(1, len(result_all)+1)
    plt.figure(figsize = (15,5))
    plt.plot(num, iv_list, label = 'qcut_iv')
    plt.title('总体IV值变化趋势')
    plt.xticks(num) #用来将横坐标的显示固定，不会自己压缩展示，但是要跟Plot语句同时运行
    plt.legend()
    plt.show()
    
    return result_all, upd
```

1.2 特征风险指标构建--基于卡方分箱

```python
#函数算法过程如下：
#1. 输入数据集，要分箱的目标变量数据（单个），标签数据， 起始的最大箱数 
#2. 基于起始箱数进行分箱，并获取箱子和标签的列联表
#3. 基于当前箱数和列联表，计算并存储woe,iv等多种指标值
#4. 基于当前列联表，根据卡方检验，合并分布最相似的两个箱子，并重置列联表
#5. 箱数减一，并判断此时箱数是否>0，如果是，返回步骤3，如果不是，结束迭代
#6. 根据得到的迭代指标结果，制作指标数据集，并输出相关图像

#补充说明：此函数是自己编写的卡方分箱计算IV值函数，功能仅限于一次处理一个特征，并且返回从最大箱数开始迭代的所有分箱结果
#只适用于连续变量，不适用于离散分类变量

def chi_bin(df, var, y, n, out_bins = 0, tp = 0):
    #df是数据集，var是目标变量；y代表'是否已经逾期'；n是初始最大箱数；tp是nan值在列联表的位置，1表示在底部，0表示在顶部，默认是0；out_bins是要显示的图像的数量,默认是0
    
    #调用需要的包
    import numpy as np
    import pandas as pd
    import scipy
    import matplotlib.pyplot as plt
    
    #如特征有缺失值，填补为'nan'
    if var.isnull().sum() > 0:
        var = var.fillna('nan') #将null填充为nan
        var[var == ''] = 'nan'
        
    #定义ks函数，以备下面使用
    def ks_2(var, y):
        from scipy.stats import ks_2samp
        var = var[var != 'nan']
        y = y[var.index]
        ks = ks_2samp(var[y == 1], var[y != 1]).statistic
        return ks
    
    #定义函数：将小数转化为百分数（字符型）
    f = lambda x: format(x, '.2%') 

    if (var == 'nan').sum() != 0: #如果变量里面有nan值，需要进行如下处理
        df['qcut'], updown = pd.qcut(var[var != 'nan'], q = n, retbins = True, duplicates = 'drop')
        df['qcut'] = np.where(df.qcut.isnull() == True, 'nan', df.qcut)
        d = pd.crosstab(df.qcut, y)
        if (tp == 0) & (d.index[0] != 'nan'): #如果nan在底部，但是要求在顶部
            d = pd.concat([d.iloc[-1:, :], d.iloc[0:-1, :]], axis = 0)
        elif (tp == 1) & (d.index[0] == 'nan'): #如果nan在顶部，但是要求在底部
            d = pd.concat([d.iloc[1:, :], d.iloc[0:1, :]], axis = 0)
        n = d.shape[0] - 1 #将n设为实际得非nan箱子个数
        flag = 1 #用于后续判断目标变量是否含有nan值
        
    else: #没有nan值得连续变量进行如下处理
        df['qcut'], updown = pd.qcut(var, q = n, retbins = True, duplicates = 'drop')
        d = pd.crosstab(df.qcut, y)
        n = d.shape[0] #分箱个数也许不是设置好的个数，保险起见，设置为实际个数
        flag = 0
        
    m = n #后面n会变化，这边先存起来，用于画图得横轴坐标
    ud = list(updown)
    is_satis_fpd, is_fpd, risk_per, woe, iv_bin, iv_bin, iv, upd, sample_rate, max_p = [], [], [], [], [], [], [], [], [], []
    while n > 0:
        is_satis_fpd_i = d.sum(axis = 1) #计算满足逾期条件的数目
        is_satis_fpd.append(is_satis_fpd_i)
        is_fpd_i = d.loc[:, 1] #计算发生逾期的次数
        is_fpd.append(is_fpd_i)
        #计算风险比率
        risk_per_i = is_fpd_i/is_satis_fpd_i
        risk_per.append(risk_per_i)
        #计算IV
        y_0_rate = (d/d.sum()).iloc[:, 0]
        y_1_rate = (d/d.sum()).iloc[:, 1]
        woe_i = np.log((y_0_rate+0.000001)/(y_1_rate+0.000001))
        iv_bin_i = ((y_0_rate - y_1_rate) * woe_i)
        iv_i = iv_bin_i.sum()
        woe.append(woe_i)
        iv_bin.append(iv_bin_i)
        iv.append(iv_i)
        #各个箱体样本比率
        rate = (d.sum(axis = 1)/d.sum(axis = 1).sum()).apply(f)
        sample_rate.append(rate)
        ud1 = ud.copy() # upd由ud添加得来，但是ud会改变，且会追溯到upd中（离谱），所以需要添加副本，以防止被改变
        upd.append(ud1)
        if n == 1 :break
        #根据卡方检验来合并分布最接近的两个箱子
        if (flag == 1) & (tp == 0):#如果是含有nan,并且要求nan放在上面，将nan箱暂时删除
            d2 = d.iloc[0:1, :]
            d = d.iloc[1:, :]
        elif (flag == 1) & (tp == 1):#如果含有nan，并且要求nan在下面
            d2 = d.iloc[-1:, :]
            d = d.iloc[0:-1, :]
        pvs = [] #用来存放卡方检验的p值
        for j in range(n-1):#遍历每个箱子，计算相邻箱子间的卡方p值
            x1 = d.iloc[j, :]
            x2 = d.iloc[j+1, :]
            pv = scipy.stats.chi2_contingency([x1+0.000001, x2+0.000001])[1] #返回卡方P值
            #chi = scipy.stats.chi2_contingency([x1, x2])[0] #返回卡方统计量值
            pvs.append(pv)
        max_p.append(max(pvs))
        max_index = pvs.index(max(pvs)) #找到最大P值对应的位置
        #先将索引处数据替换为两个箱子汇总的数据，然后在把索引+1处的数据删除，实现箱子汇总（通过拼接df数据时不带该条数据实现删除）
        d.iloc[max_index,:] = d.iloc[max_index, :] + d.iloc[max_index+1, :]
        d = pd.concat([d.iloc[:max_index+1, :], d.iloc[max_index+2:, :]], axis = 0)
        ud.pop(max_index+1) 
        d.index = [pd.Interval(ud[k], ud[k+1], closed = 'right') for k in range(len(ud)-1)] #重置索引，返回将箱子汇总后的索引
        if (flag == 1) & (tp == 0):#如果含nan，此时将汇总后得箱子再次与nan合体
            d = pd.concat([d2, d], axis = 0)
        elif (flag == 1) & (tp == 1):
            d = pd.concat([d, d2], axis = 0)
        #合并箱子后，箱数减一，继续下一次循环，直到箱数减小为0为止
        n = n - 1
        
    ks = ks_2(var, y) #引用ks_2函数计算ks值
    is_satis_fpd.reverse()
    is_fpd.reverse()
    woe.reverse()
    iv_bin.reverse()
    iv.reverse()
    risk_per.reverse()
    upd.reverse()
    sample_rate.reverse()
    risk_per_out = [] #复制risk_per,并且将数值设为百分比形式（由于百分比是字符型，在原数据设置的话无法画出正确的图）
    for i in range(len(risk_per)):
        risk_per_out.append(risk_per[i].apply(f))

    #创建输出的df
    result = []
    for i in range(len(risk_per)):
        d_fpd = pd.concat([sample_rate[i], is_satis_fpd[i], is_fpd[i], risk_per_out[i], woe[i], iv_bin[i]], axis = 1) #将三个计算字段按列拼接，合成一个df
        d_fpd.index.name = var.name + '_bin' 
        d_fpd.columns = ['各箱样本比率', '满足fpd7人数', 'fpd7人数', 'fpd7', 'woe', '各箱iv'] 
        d_iv = pd.DataFrame(index = d_fpd.index, columns = ['总体iv']).fillna(iv[i]) #创建iv字段
        d_ks = pd.DataFrame(index = d_fpd.index, columns = ['ks']).fillna(ks) #创建ks字段
        d_fpd['总体iv'], d_fpd['ks'] = d_iv, d_ks #将iv和ks字段加到d_fpd中

        if (var == 'nan').sum() != 0:
            if tp == 0: #如果nan值在顶部
                r = d_fpd.reset_index() #两个df横向相连，得到最终df
                v_name = ['nan']
                for i in range(1, r.shape[0]):#让分箱信息前面加上编号，方便看出是第几个箱子，下同
                    v_name.append('0' + str(i) + str(r.iloc[:, 0][i])) 
                r.iloc[:, 0] = v_name
                result.append(r)
            else: #如果nan值在底部
                r = d_fpd.reset_index()
                for i in range(r.shape[0]-1):
                    r.iloc[:, 0][i] = '0' + str(i+1) + str(r.iloc[:, 0][i])
                result.append(r)
        else: #如果是没有nan值的连续变量
            #涉及到pd.interval的类型数据，总是很奇怪，正常手段难以重置列，只能先删再加
            r = d_fpd.reset_index()
            new_c = ['0' + str(i+1) + str(r.iloc[:, 0][i]) for i in range(r.shape[0])]
            r.pop(var.name + '_bin')
            r.insert(0, var.name + '_bin', new_c)
            result.append(r)
    
    #将记录好的数据进行图形化展示，以便于决定最优箱数
    #这两行代码解决 plt 中文显示的问题
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    for i in range(len(risk_per)):
        if i+1 <= out_bins:
            plt.figure(figsize = (15,5))
            y_label = result[i].iloc[:, 0]
            plt.barh(y_label, risk_per[i], label = 'qcut_bin_num = ' + str(risk_per[i].shape[0]), color  ='g')  
            plt.xlabel('风险比率')
            plt.yticks(y_label)
            plt.title('各箱风险比率')
            plt.show()
            print('箱体边界值：', upd[i])
            print('各箱风险比率：',list(risk_per[i].apply(f)))
            print('各箱样本数占比：', list(sample_rate[i]))
            print('总体IV值：', iv[i])
            
    #画出总体IV值趋势图
    num = range(1, len(iv)+1)
    plt.figure(figsize = (15,5))
    plt.plot(num, iv, label = 'chi_risk_iv')
    plt.title('总体IV值变化趋势')
    plt.xticks(num) #用来将横坐标的显示固定，不会自己压缩展示，但是要跟Plot语句同时运行
    plt.legend()
    plt.show()
    
    return result, upd
```

1.3 特征风险指标构建--基于风险比率差值最小化分箱

```python
#函数的算法过程跟上面卡方分箱函数差别不大，区别在于合并箱子的规则不再是根据卡方检验，而是根据相邻箱体的fpd差值，差值最小的两个相邻箱子进行合并
#只适用于连续变量，不适用于离散分类变量
def risk_bin(df, var, y, n, out_bins = 0, tp = 0):
    #df是数据集，var是目标变量；y代表'是否已经逾期'；n是初始最大箱数；tp是nan值在列联表的位置，1表示在底部，0表示在顶部，默认是0；out_bins是输出图像的数量,默认是10
    
    #调用需要的包
    import numpy as np
    import pandas as pd
    import scipy
    import matplotlib.pyplot as plt
    
    #如特征有缺失值，填补为;nan
    if var.isnull().sum() > 0:
        var = var.fillna('nan') #将null填充为nan
        var[var == ''] = 'nan'
        
    #定义ks函数，以备下面使用
    def ks_2(var, y):
        from scipy.stats import ks_2samp
        var = var[var != 'nan']
        y = y[var.index]
        ks = ks_2samp(var[y == 1], var[y != 1]).statistic
        return ks
    
    #定义函数：将小数转化为百分数（字符型）
    f = lambda x: format(x, '.2%') 

    if (var == 'nan').sum() != 0: #如果变量里面有nan值，需要进行如下处理
        df['qcut'], updown = pd.qcut(var[var != 'nan'], q = n, retbins = True, duplicates = 'drop')
        df['qcut'] = np.where(df.qcut.isnull() == True, 'nan', df.qcut)
        d = pd.crosstab(df.qcut, y)
        if (tp == 0) & (d.index[0] != 'nan'): #如果nan在顶部，但是要求在顶部
            d = pd.concat([d.iloc[-1:, :], d.iloc[0:-1, :]], axis = 0)
        elif (tp == 1) & (d.index[0] == 'nan'): #如果nan在顶部，但是要求在底部
            d = pd.concat([d.iloc[1:, :], d.iloc[0:1, :]], axis = 0)
        n = d.shape[0] - 1 #将n设为实际得非nan箱子个数
        flag = 1 #用于后续判断目标变量是否含有nan值
        
    else: #没有nan值得连续变量进行如下处理
        df['qcut'], updown = pd.qcut(var, q = n, retbins = True, duplicates = 'drop')
        d = pd.crosstab(df.qcut, y)
        n = d.shape[0] #分箱个数也许不是设置好的个数，保险起见，设置为实际个数
        flag = 0
        
    m = n #后面n会变化，这边先存起来，用于画图得横轴坐标
    ud = list(updown)
    is_satis_fpd, is_fpd, risk_per, woe, iv_bin, iv_bin, iv, upd, sample_rate = [], [], [], [], [], [], [], [], []
    while n > 0:
        is_satis_fpd_i = d.sum(axis = 1) #计算满足逾期条件的数目
        is_satis_fpd.append(is_satis_fpd_i)
        is_fpd_i = d.loc[:, 1] #计算发生逾期的次数
        is_fpd.append(is_fpd_i)
        #计算风险比率
        risk_per_i = is_fpd_i/is_satis_fpd_i
        risk_per.append(risk_per_i)
        #计算IV
        y_0_rate = (d/d.sum()).iloc[:, 0]
        y_1_rate = (d/d.sum()).iloc[:, 1]
        woe_i = np.log((y_0_rate+0.000001)/(y_1_rate+0.000001))
        iv_bin_i = ((y_0_rate - y_1_rate) * woe_i)
        iv_i = iv_bin_i.sum()
        woe.append(woe_i)
        iv_bin.append(iv_bin_i)
        iv.append(iv_i)
        #各个箱体样本比率
        rate = (d.sum(axis = 1)/d.sum(axis = 1).sum()).apply(f)
        sample_rate.append(rate)
        ud1 = ud.copy() # upd由ud添加得来，但是ud会改变，且会追溯到upd中（离谱），所以需要添加副本，以防止被改变
        upd.append(ud1)
        if n == 1 :break
        #根据fpd7相似度最大原则合并分布最接近的两个箱子
        if (flag == 0):
            risk_per_copy = risk_per_i.copy()
        elif (flag == 1) & (tp == 0):#如果是含有nan,并且nan放在上面，将nan箱暂时删除
            d2 = d.iloc[0:1, :]
            d = d.iloc[1:, :]
            risk_per_copy = risk_per_i[1:].copy()
        elif (flag == 1) & (tp == 1):#如果含有nan，并且nan在下面
            d2 = d.iloc[-1:, :]
            d = d.iloc[0:-1, :]
            risk_per_copy = risk_per_i[0:-1].copy()
        risk_diff_n = [] #用来存放相邻箱体风险比率的差值
        for j in range(n-1):#遍历每个箱子，计算相邻箱子间的风险比率差值
            x1 = risk_per_copy.values[j]
            x2 = risk_per_copy.values[j+1]
            risk_diff_n.append(abs(x1 - x2))
        min_index = risk_diff_n.index(min(risk_diff_n)) #找到最小风险比率差值对应的位置
        #先将索引处数据替换为两个箱子汇总的数据，然后在把索引+1处的数据删除，实现箱子汇总（通过拼接df数据时不带该条数据实现删除）
        d.iloc[min_index,:] = d.iloc[min_index, :] + d.iloc[min_index+1, :]
        d = pd.concat([d.iloc[:min_index+1, :], d.iloc[min_index+2:, :]], axis = 0)
        ud.pop(min_index+1) 
        d.index = [pd.Interval(ud[k], ud[k+1], closed = 'right') for k in range(len(ud)-1)] #重置索引，返回将箱子汇总后的索引
        if (flag == 1) & (tp == 0):#如果含nan，此时将汇总后得箱子再次与nan合体
            d = pd.concat([d2, d], axis = 0)
        elif (flag == 1) & (tp == 1):
            d = pd.concat([d, d2], axis = 0)
        #print(d)
        #合并箱子后，箱数减一，继续下一次循环，直到箱数减小为0为止
        n = n - 1
        
    ks = ks_2(var, y) #引用ks_2函数计算ks值
    is_satis_fpd.reverse()
    is_fpd.reverse()
    woe.reverse()
    iv_bin.reverse()
    iv.reverse()
    risk_per.reverse()
    upd.reverse()
    sample_rate.reverse()
    risk_per_out = [] #复制risk_per,并且将数值设为百分比形式（由于百分比是字符型，在原数据设置的话无法画出正确的图）
    for i in range(len(risk_per)):
        risk_per_out.append(risk_per[i].apply(f))

    #创建输出的df
    result = []
    for i in range(len(risk_per)):
        d_fpd = pd.concat([sample_rate[i], is_satis_fpd[i], is_fpd[i], risk_per_out[i], woe[i], iv_bin[i]], axis = 1) #将三个计算字段按列拼接，合成一个df
        d_fpd.index.name = var.name + '_bin' 
        d_fpd.columns = ['各箱样本比率', '满足fpd7人数', 'fpd7人数', 'fpd7', 'woe', '各箱iv'] 
        d_iv = pd.DataFrame(index = d_fpd.index, columns = ['总体iv']).fillna(iv[i]) #创建iv字段
        d_ks = pd.DataFrame(index = d_fpd.index, columns = ['ks']).fillna(ks) #创建ks字段
        d_fpd['总体iv'], d_fpd['ks'] = d_iv, d_ks #将iv和ks字段加到d_fpd中

        if (var == 'nan').sum() != 0:
            if tp == 0: #如果nan值在顶部
                r = d_fpd.reset_index() #两个df横向相连，得到最终df
                v_name = ['nan']
                for i in range(1, r.shape[0]):#让分箱信息前面加上编号，方便看出是第几个箱子，下同
                    v_name.append('0' + str(i) + str(r.iloc[:, 0][i])) 
                r.iloc[:, 0] = v_name
                result.append(r)
            else: #如果nan值在底部
                r = d_fpd.reset_index()
                for i in range(r.shape[0]-1):
                    r.iloc[:, 0][i] = '0' + str(i+1) + str(r.iloc[:, 0][i])
                result.append(r)
        else: #如果是没有nan值的连续变量
            #涉及到pd.interval的类型数据，总是很奇怪，正常手段难以重置列，只能先删再加
            r = d_fpd.reset_index()
            new_c = ['0' + str(i+1) + str(r.iloc[:, 0][i]) for i in range(r.shape[0])]
            r.pop(var.name + '_bin')
            r.insert(0, var.name + '_bin', new_c)
            result.append(r)
    
    #将记录好的风险数据进行图形化展示，以便于决定最优箱数
    #这两行代码解决 plt 中文显示的问题
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    for i in range(len(risk_per)):
        if i+1 <= out_bins:
            plt.figure(figsize = (15,5))
            y_label = result[i].iloc[:, 0]
            plt.barh(y_label, risk_per[i], label = 'qcut_bin_num = ' + str(risk_per[i].shape[0]), color  ='g')  
            plt.xlabel('风险比率')
            plt.yticks(y_label)
            plt.title('各箱风险比率')
            plt.show()
            print('箱体边界值：', upd[i])
            print('各箱风险比率：',list(risk_per[i].apply(f)))
            print('各箱样本数占比：', list(sample_rate[i]))
            print('总体IV值：', iv[i])
    
    #画出总体IV值趋势图
    num = range(1, len(iv)+1)
    plt.figure(figsize = (15,5))
    plt.plot(num, iv, label = 'chi_risk_iv')
    plt.title('总体IV值变化趋势')
    plt.xticks(num) #用来将横坐标的显示固定，不会自己压缩展示，但是要跟Plot语句同时运行
    plt.legend()
    plt.show()
            
    return result, upd
```

1.4 KS值计算函数介绍

```python
#计算KS值得函数已经在上述函数内部进行编写调用，在这里是为了进行详细得说明
#只适用于连续型变量,只能使用pandas类型数据，因为里面加了只针对pandas类型的去除缺失值操作
def ks_1(var, y): #y是标签，var是目标变量
    from sklearn import metrics
    var = var[var != 'nan']
    y = y[var.index]
    fpr, tpr, thresholds = metrics.roc_curve(y, var, pos_label=1, drop_intermediate=False) #最后一个参数的作用在于不会省略某些阈值
    return max(abs(tpr-fpr)) #由于不确定var的值对标签的影响方向，所以很可能出现本该var越小标签越倾向于1，但是实际执行却是相反，这样算出的ks值是正确值的相反数，所以需要对ks值加abs取绝对值，以确保计算正确

#KS值-2
def ks_2(var, y):
    from scipy.stats import ks_2samp
    var = var[var != 'nan']
    y = y[var.index]
    ks = ks_2samp(var[y == 1], var[y != 1]).statistic
    return ks
```

3.3.4 数据编码

```python
# 由于数据集全是分箱（或分类）变量（sklearn中决策树只能处理数值型）
# 这里根据上文，这里只针对分箱后的特征操作，所以全部需要编码，但由于原始连续型数据不需要编码，只有分箱或者原始分类特征数据才需要编码，所以如果使用原始特征跑模型，对数据是否进行编码，对哪些特征编码需要自己选择
from category_encoders import CatBoostEncoder, TargetEncoder, WOEEncoder #引入特征编码的包

x_train, x_test, y_train, y_test = train_test_split(df2.iloc[:, :-1], df2.iloc[:, -1], test_size=0.3, random_state=0) #拆分训练集和测试集
var_list = list(x_train.columns[0:1]) + list(x_train.columns[25:]) #需要编码的特征名称

#然后进行WOE编码
# handle的两个参数选项是 'error'、'return_nan' 和 'value'，默认为 'value'，对unknown(新出现的特征值)和missing(缺失值)的woe设置为0
woe_encoder = WOEEncoder(cols = var_list, handle_unknown='value', handle_missing='value').fit(x_train, y_train)
woe_encoder_x_train = woe_encoder.transform(x_train) # 转换训练集
woe_encoder_x_test = woe_encoder.transform(x_test) # 转换测试集

```

```python
# 对于分箱数据，一般被编码为woe等值，跑出来的结果也都以woe（或其他编码值）形式存在的，我们不能直观得获取特征信息，所以需要建立一个表用来对应每个特征的编码值跟其所属分箱（或分组）
#以woe编码为例，编写编码对照表，用于给分箱数据的结果（woe形式）跟其规则形成对应
woe_encoder_x_train_dict = pd.DataFrame(columns = ['name', 'bins', 'woe', 'count']) #创建编码前训练集的df
for i in var_list:
    name_bin = [i] * (x_train[i].value_counts().shape[0]) #统计出编码前训练集特征名
    bins_bin = x_train[i].value_counts().sort_values(ascending = False).index.to_list() #统计出编码前数据集特征分箱情况或类别名称，并按照频数排序
    sample_num_bin = x_train[i].value_counts().sort_values(ascending = False).values #统计编码前数据集分箱或分类后每组特征值频数，并排序
    woe_bin = woe_encoder_x_train[i].value_counts().sort_values(ascending = False).index.to_list() #统计编码后数据集的woe值，并进行分组排序（编码后数据集是对）
    d_bin = pd.DataFrame({'name':name_bin, 'bins':bins_bin, 'count':sample_num_bin}) #将通过编码前训练集得到的列合成df类型
    woe_bin = pd.DataFrame(woe_bin, columns = ['woe']) #将根据编码后训练集得到的列也转为df类型
    d_bin = pd.concat([d_bin, woe_bin], axis = 1) #将两个df合并
    woe_encoder_x_train_dict = woe_encoder_x_train_dict.append(d_bin) #不同特征的结果合并起来
df_encode_dict = woe_encoder_x_train_dict[['name', 'bin', 'woe', 'count']] #最终的精简结果对照表
```

4.5.4 创建分层lift表

```python
# 决策树做好后，我们可以追求模型效果，也可以根据特征重要性指标获取特征的作用排名，但更加常见的业务需求是根据决策树做规则，使用规则分层，指南上会有详情，这里就根据所得决策树进行分层创建lift表

# 决策树规则拆解算法：
# 1. 编写分层函数——将决策树拆解成一个个规则，并以多维列表的形式存放
#   规则列表（假设为a）有三个维度，第一维度是规则本身，比如a[0]代表第一条规则；第二个维度是此规则在每个深度的节点做出的选择，如a[0][0]代表第一条规则在第一个深度节点的选择；
#   第三个维度是节点选择的具体内容，包括特征，连接方式和特征阈值，如a[0][0][0]代表第一条规则在第一个节点的特征，a[0][0][1]是'<='或者'>'，a[0][0][2]代表阈值
#   如第一条规则在第一个深度对应节点的分支规则是特征b<=10，则a[0][0][0]=b, a[0][0][1]='<=', a[0][0][2]=10
# 2. 然后根据决策树得最大深度，穷尽所有可能得走法，以[0,1,1,0,1]这种形式存储每个规则（分支）得全部走法，0代表向左走，1代表向右走，位置代表所处深度
# 3. 根据每个规则的走法，读取决策树按照这种走法获取的规则内容，并以多维列表的形式进行保存，重复过程直到所有规则走完
# 4. 注意：对于那些没达到最大深度就结束得分支，需要进行判断，判断其分支结束就停止继续走下去
def tree_rule(model, feature_names): #输入参数:model是拟合了数据集后的决策树模型，feature_names是数据集中特征名称
    tree_ = model.tree_
    feature_name = [feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!" for i in tree_.feature]

    def recurse(node, depth, flag):
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            if flag == '0':
                return [name, '<=', threshold]
            else:
                return [name, '>', threshold]
    
    max_depth = model.get_params()['max_depth']
    bin_list = []
    for i in range(pow(2, max_depth)):
        bin_list.append(list(bin(i)[2:].rjust(max_depth,'0')))
    rule_all_list = []
    for j in range(pow(2, max_depth)):
        rule_list = []
        node = 0
        flag = bin_list[j][0]
        for k in range(max_depth):
            if k == 0:
                rule_list.append(recurse(node, k+1, flag))
            else:
                if flag == '0':
                    node = tree_.children_left[node]
                    if tree_.children_left[node] != -1:
                        flag = bin_list[j][k]
                        rule_list.append(recurse(node, k+1, flag))
                    else:
                        break
                else:
                    node = tree_.children_right[node]
                    if tree_.children_right[node] != -1:
                        flag = bin_list[j][k]
                        rule_list.append(recurse(node, k+1, flag))
                    else:
                        break
        rule_all_list.append(rule_list)

    unique_list = [rule_all_list[0]]
    for i in range(len(rule_all_list)-1):
        if rule_all_list[i] != rule_all_list[i+1]:
            unique_list.append(rule_all_list[i+1])
    
    return unique_list

#编写生成lift表函数，需要用到上个函数得到的规则列表
#lif表包括训练集和测试集的结果，将他们进行横向拼接得到df类型数据，并且按照训练集的fpd7进行升序排序
def df_lift(data, data1, rule_list): #输入参数：data是训练数据集，就是训练决策树的，data1是测试数据集，rule_list是规则列表
    f = lambda x: format(x, '.2%') #将小数字转化为百分比函数
    data_split_lift = pd.DataFrame([[data.shape[0], 1, data.is_fpd7.sum(), data.is_fpd7.mean(), 1]], index = ['all_data'], 
                                 columns = ['训练集总人数', '训练集人数占比', '训练集fpd7人数', '训练集fpd7', '训练集lift值'])
    data_split_lift1 = pd.DataFrame([[data1.shape[0], 1, data1.is_fpd7.sum(), data1.is_fpd7.mean(), data1.is_fpd7.mean()/data.is_fpd7.mean()]], index = ['all_data'], 
                             columns = ['测试集总人数', '测试集人数占比', '测试集fpd7人数', '测试集fpd7', '测试集lift值'])
    num = 1
    for i in rule_list:
        d =  data.copy()
        d1 =  data1.copy()
        for j in i:
            if j[1] == '<=':
                d = d[d[j[0]] <= j[2]]
                d1 = d1[d1[j[0]] <= j[2]]
            else:
                d = d[d[j[0]] > j[2]]
                d1 = d1[d1[j[0]] > j[2]]
        print(d)
        count = d.shape[0]
        count1 = d1.shape[0]
        count_per = count/data.shape[0]
        count_per1 = count1/data1.shape[0]
        fpd7_cnt = d.iloc[:, -1].sum()
        fpd7_cnt1 = d1.iloc[:, -1].sum()
        fpd7 = d.iloc[:, -1].mean()
        fpd71 = d1.iloc[:, -1].mean()
        lift = fpd7/data.iloc[:, -1].mean()
        lift1 = fpd71/data.iloc[:, -1].mean()
        d_split_lift = pd.DataFrame([[count, count_per, fpd7_cnt, fpd7, lift]], index = ['rule_'+str(num)], 
                                 columns = ['训练集总人数', '训练集人数占比', '训练集fpd7人数', '训练集fpd7', '训练集lift值'])
        d_split_lift1 = pd.DataFrame([[count1, count_per1, fpd7_cnt1, fpd71, lift1]], index = ['rule_'+str(num)], 
                                 columns = ['测试集总人数', '测试集人数占比', '测试集fpd7人数', '测试集fpd7', '测试集lift值'])
        data_split_lift = data_split_lift.append(d_split_lift)
        data_split_lift1 = data_split_lift1.append(d_split_lift1)
        num = num + 1

    data_lift = pd.concat([data_split_lift, data_split_lift1], axis = 1)
    data_lift_rule = data_lift.iloc[1:, :].sort_values(by = '训练集fpd7')
    data_lift_all = data_lift.iloc[:1, :]
    data_lift = pd.concat([data_lift_all, data_lift_rule], axis = 0)
    data_lift['测试集人数占比'] = data_lift['测试集人数占比'].apply(f)
    data_lift['训练集人数占比'] = data_lift['训练集人数占比'].apply(f)
    data_lift['测试集fpd7'] = data_lift['测试集fpd7'].apply(f)
    data_lift['训练集fpd7'] = data_lift['训练集fpd7'].apply(f)
    return data_lift

#此函数使用场景：想要将满足几条规则的样本筛选出来，对其进行进一步操作
#输入数据集和规则列表，以及需要使用的规则序号（如第1/3/5三条规则），自动输出满足目标规则的数据集
#注意：目标规则需要的输入方式，如[1,3,5]:代表第1/3/5三条规则
def df_rule(data, rule_list, rule_num_list):
    r_list = []
    for i in rule_num_list:
        r_list.append(rule_list[i-1])
    df = pd.DataFrame(columns = data.columns)
    for j in r_list:
        d = data.copy()
        for k in j:
            if k[1] == '<=':
                d = d[d[k[0]] <= k[2]]
            else:
                d = d[d[k[0]] > k[2]]
        df = df.append(d)
    return df

rule_list = tree_rule(model, x1.columns) #拆解决策树得到规则

# 分层效果校验——测试集效果检验
data = pd.concat([x1, y1], axis = 1) #训练集数据
data1 = pd.concat([x2, y2], axis = 1) #测试集数据
data_lift = df_lift(data, data1, rule_list) #获取测试集lift表
data_rule = df_rule(data, rule_list, [1]) #取出满足某些规则的数据集，一种方法而已，看自己需求使用

# 分层效果校验——逐月效果检验
# 获取所有授信年月信息，去重后升序排成列表
date_list = sorted(list(set(df.credit_submit_date_month.sort_values().tolist())))
month_list = []
for i in date_list:
    month_list.append(i[:7])
month_list = sorted(list(set(month_list)))

# 将训练集和测试集数据集合在一起，获得总数据，并将数据按照时间顺序按月切分数据，分别存储再data_month中（这里使用授信月作为时间维度：credit_submit_date_month，实际使用灵活选择）
# 轮流将每个月得数据作为测试集创建lift表，并依次存储在data_lift_month中
data_all = data.append(data1) #合并为总数据
data_month = [] #创建列表用来存储每个月的数据集
data_lift_month = [] # 创建列表存储每个月的lift检验表
# 下面是创建分月数据和相应lift表的过程（注：可以不以单个月份为单位，选择三个月/半年等均可，只需要在下面的代码中对循环的步长和判断时间大小时i的选择做更改即可）
for i in range(len(month_list)):
    data_month.append(data_all[(df.credit_submit_date_month >= month_list[i] + '-01') & (df.credit_submit_date_month <=  month_list[i] +'-31')]) #创建分月数据
    data_lift_month.append(df_lift(data, data_month[-1], rule_list)) #对分月数据创建Lift表，然后存储
```
