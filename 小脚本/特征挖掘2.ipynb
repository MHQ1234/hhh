2 单变量挖掘过程

2.1 读取处理好得数据集(如无修改，直接读取，不需追溯前面代码）

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import os 
import glob
import seaborn as sn
from sklearn.model_selection import train_test_split
import pickle
#设置显示格式
pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows', 500)
#开始读取保存好的变量
f = open("./样本筛选后的数据集.txt",'rb')
df = pickle.load(f)
f.close() #读取完毕
```

2.2 筛选备用特征，并进行分箱

```python
import scorecardpy as sc #评分卡包
import re #正则表达式模块

# 首先是选出自己想要分析的连续型特征，因为这些特征需要分箱，将他们的名称放在var_bin_list列表中，（如果全都是要分析的连续型特征，那么全部加进来），将那些一定不会使用的特征筛出去，这里并无统一规则，我是按照特征名称的规则使用正则表达式筛选
# 而对于需要分析的分类型特征，且并不想将他们合并的话，先不管它们，后面会将分箱后的结果与这些分类特征合并
# 最终创建出的数据集是所有备选分析特征加上标签
var_bin_list = [] #备选连续型特征名称存放列表
for i in range(df.shape[1]):
    text = df.columns[i]
    pattern = r'last\w+|\w+flag|reciprocal\w+' #正则表达式，用于得到满足条件的特征名称
    f = re.findall(pattern, text) #返回判别结果，注意：返回的是列表型，里面存放着字符串
    if len(f) != 0:
        var_bin_list.extend(f) #将满足条件得特征名称存放起来，添加列表型里面的内容需要使用extend函数
var_bin_list.extend(['is_fpd7']) #再将标签特征加到数据集中
df1 = df.loc[:, var_bin_list] #创建新的数据集
var_list_corr= list(df1.columns[:-1]) #提取需要分箱的特征列表


#对筛选后的数据集进行分箱
# 这里使用到了业界大佬编写的评分卡库scorecardpy，这个包内容很丰富，可以实现批量分箱子（自行设置分箱方法，分箱细节参数）
# 这里用到了这个库下面的两个函数，一个是下面马上使用的woebin(),另一个是后面用到的woebin_ply()
# 关于函数的介绍，内容较多，请点此链接查看：https://blog.csdn.net/qq_41144446/article/details/107839967
# tips:这个函数不仅会对连续型变量进行分箱，也会对分类变量进行合并压缩，但当分类特征的类别不是很多时，我们不希望其对分类变量进行合并，只需其保持原有类别，一类代表一个分组即可，
#      这个时候建议仅在var_list_corr中放置需要分箱的特征名称，对于不需要分箱的特征，我们最后将其加入到分箱结果中即可
# 虽然分类离散型特征可能不需要分箱，但他们也需要查看IV，以进行可能的筛选，可用另一个文件里面的qcut_bin函数进行风控指标的建立
bins_app = sc.woebin(df1, y="is_fpd7", x=var_list_corr, positive = "bad|1", count_distr_limit=0.01, stop_limit=0.5, print_step=1, method='tree',
                     ignore_const_cols=True, ignore_datetime_cols=True, check_cate_num=True, replace_blank=True, max_num_bin=10)


# 以上结果使用字典型存储，特征为键，结果df为内容。在这里我们将结果更加规范化，从字典型转为df型拼接
chimerge_bins_app=pd.DataFrame(columns=('variable','bin','count','count_distr','good','bad','badprob','woe','bin_iv','total_iv','breaks','is_special_values'))
for i in bins_app.keys():
    data_app=pd.DataFrame(bins_app[i])
    chimerge_bins_app=chimerge_bins_app.append(data_app) # chimerge_bins_app即为规范化输出结果
chimerge_bins_app = chimerge_bins_app.sort_values(by = 'total_iv', ascending = False) #将特征按照总IV值降序排序
chimerge_bins_app.to_csv('./决策树分箱结果.csv', index = True, encoding="utf_8_sig") #存储对连续型特征分箱结果为csv文件
```

2.3 根据分箱结果，对特征进行筛选，并获取跑模型原始数据集

```python
#特征筛选及存储
# 分箱结束后，我们需要自行查看每个特征的分箱结果，和特征的风险比率是否有比较好的排序性，然后选择自觉合适的特征准备跑模型
# 这个过程并没有统一标准，只要按照自己的理解选择即可，我这里是简单的将IV小于0.02特征删掉，完全可以自己在execl上选出排序性很好，IV和KS也不错的特征，只要最后将这些特征选出来即可
# 如果当时不想对某些分类变量分箱，暂时剔除，这时就需要将其加入到最终的特征名称列表中
# 总之最后的筛选出来的连续型特征和一部分分类特征，要进行结合，并加上标签，共同将他们的名称存放在var_list列表中
chimerge_bins_app = chimerge_bins_app[chimerge_bins_app.total_iv >= 0.02] #选出IV不低于0.02的特征
var_list = list(set(list(chimerge_bins_app.variable))) #获取入选特征名称
var_list.extend(['driver_active', 'lifecycle1','is_fpd7']) #加上两个分类特征名以及标签名

#在使用IV/ks/风险排序型等指标对分箱后的特征进行筛选之后，需要使用入选特征的数据集跑模型，但是使用原始数据信息还是分箱后的信息呢？
# 建议做两版，原始信息和分箱后的信息分别跑模型，得到两版结果做对比，首先采用原始信息，并对其进行简单缺失值处理
data_set = df.loc[:, var_list] #从原始数据集中获取需要的特征的数据
data_set = pd.DataFrame(np.where(data_set.isnull() == True, -9999, data_set), columns = data_set.columns) #将所有缺失值填补为-9999，因为sklearn跑机器学习模型只能处理数值型
data_set['is_fpd7'] = data_set.is_fpd7.astype(np.int) #把可能是字符型数字的标签数值化处理
data_set.to_csv('./跑模型原始数据集.csv', index = False, encoding="utf_8_sig")
```

2.4 对原始数据进行编码

```python
# 从上面的信息可以得知，特征中可能存在并不需要被分箱的分类型特征，对于这些特征，不能直接在py中跑模型，因为sklearn机器学习库里面的绝大多数模型只能处理数值型特征
# 所以需要对所有字符型特征进行编码，将其转化为数值型特征，对于特征编码，方法和内容都很多，详见链接：https://mattzheng.blog.csdn.net/article/details/107851162
# 这里我们使用woe编码，此编码更符合风控习惯

#跑原始数据仅需要对类别特征编码
from category_encoders import WOEEncoder #引入特征编码的包
var_encode_list = [] #需要编码的名称存放列表
# 将需要编码的特征名称存放到这个列表，方法不限，不一定要用下面的
for i in range(data_set.shape[1]):
    text = data_set.columns[i]
    pattern = r'life\w+|\w+bin|driver_active' #正则表达式，用于得到满足条件的特征名称
    f = re.findall(pattern, text) #返回判别结果，注意：返回的是列表型，里面存放着字符串
    if len(f) != 0:
        var_encode_list.extend(f) #将满足条件得特征名称存放起来，添加列表型里面的内容需要使用extend函数

        
# 编码前需要把数据拆分为训练集和测试集，对训练集中需要编码的特征进行woe编码，然后对于测试集，根据训练集的编码结果去对应编码即可
# 这样做的原因是，woe编码是用到了现有样本的所有标签信息的，我们把测试集视为未知数据集，所以不能使用它的标签信息编码然后再预测它的标签，这样会过拟合，所以先用训练集编码，然后测试集的编码只需要按照与训练集相对应的值进行编码即可
# 对于测试集，找到其值对应的训练集中相同的值，然后使用训练集中此值的编码即可
x_train, x_test, y_train, y_test = train_test_split(data_set.iloc[:, :-1], data_set.iloc[:, -1], test_size=0.3, random_state=0) #拆分训练集和测试集

#然后进行WOE编码
# handle的两个参数选项是 'error'、'return_nan' 和 'value'，默认为 'value'，对unknown(新出现的特征值)和missing(缺失值)的woe设置为0
woe_encoder = WOEEncoder(cols = var_encode_list, handle_unknown='value', handle_missing='value').fit(x_train, y_train) #对训练集合的分类特征进行woe编码
woe_encoder_x_train = woe_encoder.transform(x_train) # 转换训练集
woe_encoder_x_test = woe_encoder.transform(x_test) # 转换测试集


#分类（或分箱）特征及其woe编码对照表
# 意义：对于分箱数据，一般被编码为woe等值，跑出来的结果也都以woe（或其他编码值）形式存在的，我们不能直观得获取特征信息，所以需要建立一个表用来对应每个特征的编码值跟其所属分箱（或分组）
# 原始数据集woe值与分类信息对照原理：
# 1. 原始数据集中只有分类数据需要编码，所以只需要对照分类数据，连续型数据的缺失也都填充为了-9999，不需要处理
# 2. 从编码前的数据中进行分组统计，得到分类特征名，分组所有类别以及对应频数，并按照频数进行排序，然后对编码后数据集的woe值进行同样操作，为实现一一对应
# 3.将从编码前后数据集中得到的列分别建立df，并合并在一起，由于都是按照频数排序的，所以基本可以实现一一对应
# （分开创建的原因是：编码后woe值可能出现不同分组的woe值相同导致统计的时候和合成了一个组，从而列长度出现不匹配，无法创建在一个df中，分开创建再合并的话，不匹配也可以拼在一起，出现NAN而已）
woe_encoder_x_train_dict = pd.DataFrame(columns = ['name', 'bins', 'woe', 'sample_num'])
for i in var_encode_list:
    name = [i] * (x_train[i].value_counts().shape[0]) #统计出编码前训练集特征名
    bins = x_train[i].value_counts().sort_values(ascending = False).index.to_list() #统计出编码前数据集特征分箱情况或类别名称，并按照频数排序
    sample_num = x_train[i].value_counts().sort_values(ascending = False).values #统计编码前数据集分箱或分类后每组特征值频数，并排序
    woe = woe_encoder_x_train[i].value_counts().sort_values(ascending = False).index.to_list() #统计编码后数据集的woe值，并进行分组排序
    d = pd.DataFrame({'name':name, 'bin':bins, 'count':sample_num}) #将通过编码前训练集得到的列合成df类型
    woe = pd.DataFrame(woe, columns = ['woe']) #将根据编码后训练集得到的列也转为df类型
    d = pd.concat([d, woe], axis = 1) #将两个df合并
    woe_encoder_x_train_dict = woe_encoder_x_train_dict.append(d) #不同特征的结果合并起来
df_encode_dict = woe_encoder_x_train_dict[['name', 'bin', 'woe', 'count']] #挑选要显示的字段
```

3. 获取目标特征的分箱结果数据集（用到的变量信息都是上一段原始数据集的变量信息）

```python
# 处理完原始数据集，现在处理分箱数据集，为第二版结果准备数据集
# 对于分箱数据集，既有连续型变特征分箱的结果，也有选择不合并的分类特征，所有这写特征都是字符型，都需要进行编码
# 前面我们得到的bins_app变量是分箱的结果呈现，我们可以使用下面这个函数woebin_ply，直接将其转化为woe编码结果数据集
# 关于函数的介绍，内容较多，请点此链接查看：https://blog.csdn.net/qq_41144446/article/details/107839967
train_woe = sc.woebin_ply(df1.loc[:, var_list], bins_app) #var_list是上面2.3部分筛选后的特征名称列表，将这些特征的分箱结果进行编码
train_woe.index = range(train_woe.shape[0])

#将分类特征添加或覆盖到编码后的数据集中
# train_woe作为编码结果，其内部特征名称都被加了_woe后缀，这里
var_encode_list_woe = []
for i in var_encode_list: #分类特征名称一一对应
    var_encode_list_woe.append(i + '_woe')
for i in range(len(var_encode_list_woe)): #添加或替换分类特征数据
    train_woe.loc[:, var_encode_list_woe[i]] = data_set.loc[:, var_encode_list[i]]
train_woe = train_woe.loc[:, (train_woe.columns != 'lifecycle1') & (train_woe.columns != 'driver_active')] #剔除两个无用特征（已被其他名字的相同特征取代）
```

4. 对分箱结果数据集进行编码

```python
# 这部分有点乱且有瑕疵，本来编码是对分箱信息进行编码，但woebin_ply输出的直接是编码后的结果，所以得到的数据是编码后的原始数据+分类数据，要对它们拆分训练集和测试集之后再编码
# 此时原woe编码可视为分箱信息，因为woe与分箱是对应得，但瑕疵在于woe编码可能会重复
# 不过分追求严谨的话，可以直接用连续型特征得第一遍得编码作为最终编码，只对分类特征编码然后拼到一起，拆分数据集直接使用即可

var_encode_bin_list = list(train_woe.columns[1:]) #需要被编码特征名称列表，包括所有分箱特征，将已被编码的特征的woe值视为箱子代号进行重新编码
x_bin_train, x_bin_test, y_bin_train, y_bin_test = train_test_split(train_woe.iloc[:, 1:], train_woe.iloc[:, 0], test_size=0.3, random_state=0) #拆分训练集和测试集

#然后进行WOE编码
# handle的两个参数选项是 'error'、'return_nan' 和 'value'，默认为 'value'，对unknown(新出现的特征值)和missing(缺失值)的woe设置为0
woe_encoder_bin = WOEEncoder(cols = var_encode_bin_list, handle_unknown='value', handle_missing='value').fit(x_bin_train, y_bin_train) #对训练集合的所有特征进行woe编码
woe_encoder_x_bin_train = woe_encoder_bin.transform(x_bin_train) # 转换训练集
woe_encoder_x_bin_test = woe_encoder_bin.transform(x_bin_test) # 转换测试集

train_encode_bin_set = pd.concat([woe_encoder_x_bin_train, y_bin_train], axis = 1)
train_encode_bin_set.to_csv('./train_encode_bin_set.csv', index = False, encoding="utf_8_sig")
test_encode_bin_set = pd.concat([woe_encoder_x_bin_test, y_bin_test], axis = 1)
test_encode_bin_set.to_csv('./test_encode_bin_set.csv', index = False, encoding="utf_8_sig")

#创建分类（或分箱）特征及其woe编码对照表
# woe值和分箱信息对照原理：
# 0.使用woebin函数只对连续型特征编码，并将编码值视为分箱箱体，将其与分类特征拼在一起，形成跑模型的数据集，并拆分出训练集和测试集，然后对训练集所有特征再次编码
# 1.分箱数据的编码前训练集包括编过一次码（woebin函数编码）的连续型特征和没编码的分类特征，编码后训练集是对连续特征二次编码，对分类特征一次编码的结果
# 2.从编码前的数据中进行分组统计，得到特征名，分组所有类别以及对应频数，并进行排序，然后对编码后数据集的woe值进行同样操作
# 3.将从编码前后数据集中得到的列分别建立df，并合并在一起，由于都是按照频数排序的，所以基本可以实现一一对应
# （分开创建的原因是：编码后woe值可能出现不同分组的woe值相同导致统计的时候和合成了一个组，从而列长度出现不匹配，无法创建在一个df中，分开创建再合并的话，不匹配也可以拼在一起，出现NAN而已）
# 4.此时已经得到分箱数据训练集的特征分箱（或分组）和woe值的对应情况，但是此事所有连续型特征的分箱（或分组）都是经过了woebin函数编码一次的woe值结果，并非真实箱子
# 5.将这个df与bins_app进行类似于sql的表连接，因为bins_app中存储了连续型特征的一次编码值对应的原始分箱情况，将df的分箱列于bins_app中的woe值列作为连接条件，以df为左连接主表
#  这样可以把df中的连续特征分箱情况（实际是编码值）对应到其真正的原始分箱，而df中的分类特征也不会再连接中被丢掉
woe_encoder_x_bin_train_dict = pd.DataFrame(columns = ['name', 'bins', 'woe_train', 'count']) #创建编码前训练集的df
for i in var_encode_bin_list:
    name_bin = [i] * (x_bin_train[i].value_counts().shape[0]) #统计出编码前训练集特征名
    bins_bin = x_bin_train[i].value_counts().sort_values(ascending = False).index.to_list() #统计出编码前数据集特征分箱情况或类别名称，并按照频数排序
    sample_num_bin = x_bin_train[i].value_counts().sort_values(ascending = False).values #统计编码前数据集分箱或分类后每组特征值频数，并排序
    woe_bin = woe_encoder_x_bin_train[i].value_counts().sort_values(ascending = False).index.to_list() #统计编码后数据集的woe值，并进行分组排序（编码后数据集是对）
    d_bin = pd.DataFrame({'name':name_bin, 'bins':bins_bin, 'count':sample_num_bin}) #将通过编码前训练集得到的列合成df类型
    woe_bin = pd.DataFrame(woe_bin, columns = ['woe_train']) #将根据编码后训练集得到的列也转为df类型
    d_bin = pd.concat([d_bin, woe_bin], axis = 1) #将两个df合并
    woe_encoder_x_bin_train_dict = woe_encoder_x_bin_train_dict.append(d_bin) #不同特征的结果合并起来
    
#特征名称操作，去除名称中的_woe后缀，以备下面进行表连接操作
name_list = woe_encoder_x_bin_train_dict.name.to_list()
var_new_bin_list = []
for i in name_list:
    var_new_bin_list.append(''.join(list(i)[:-4]))
woe_encoder_x_bin_train_dict['name'] = var_new_bin_list

#将df与bins_app进行表连接，连接方式：df左连接；连接条件：二者特征名称是第一个条件，df的分箱bins连接bins_app的woe值
df_bin_encode_dict = pd.merge(woe_encoder_x_bin_train_dict, chimerge_bins_app[['variable', 'woe', 'bin']], how = 'left', left_on = ['name', 'bins'], right_on = ['variable', 'woe'])
#将df和bins_app的分箱列进行合并统一，所有连续型特征分箱使用bins_app的列，所有分类型特征值使用df的列
df_bin_encode_dict['bin'] = np.where(df_bin_encode_dict.bin.isnull() == True, df_bin_encode_dict.bins, df_bin_encode_dict.bin) 
df_bin_encode_dict = df_bin_encode_dict[['name', 'bin', 'woe_train', 'count']] #挑选要显示的字段
df_bin_encode_dict = df_bin_encode_dict.sort_values(by = ['name', 'woe_train'], ascending = False)
df_bin_encode_dict.to_csv('./分箱数据参照表.csv', index = False, encoding="utf_8_sig")
```

5.3. 创建分层lift表

```python
# 决策树做好后，我们可以追求模型效果，也可以根据特征重要性指标获取特征的作用排名，但更加常见的业务需求是根据决策树做规则，使用规则分层，指南上会有详情，这里就根据所得决策树进行分层创建lift表

# 决策树规则拆解算法：
# 1. 编写分层函数——将决策树拆解成一个个规则，并以多维列表的形式存放
#   规则列表（假设为a）有三个维度，第一维度是规则本身，比如a[0]代表第一条规则；第二个维度是此规则在每个深度的节点做出的选择，如a[0][0]代表第一条规则在第一个深度节点的选择；
#   第三个维度是节点选择的具体内容，包括特征，连接方式和特征阈值，如a[0][0][0]代表第一条规则在第一个节点的特征，a[0][0][1]是'<='或者'>'，a[0][0][2]代表阈值
#   如第一条规则在第一个深度对应节点的分支规则是特征b<=10，则a[0][0][0]=b, a[0][0][1]='<=', a[0][0][2]=10
# 2. 然后根据决策树得最大深度，穷尽所有可能得走法，以[0,1,1,0,1]这种形式存储每个规则（分支）得全部走法，0代表向左走，1代表向右走，位置代表所处深度
# 3. 根据每个规则的走法，读取决策树按照这种走法获取的规则内容，并以多维列表的形式进行保存，重复过程直到所有规则走完
# 4. 注意：对于那些没达到最大深度就结束得分支，需要进行判断，判断其分支结束就停止继续走下去
def tree_rule(model, feature_names): #输入参数:model是拟合了数据集后的决策树模型，feature_names是数据集中特征名称
    tree_ = model.tree_
    feature_name = [feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!" for i in tree_.feature]

    def recurse(node, depth, flag):
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            if flag == '0':
                return [name, '<=', threshold]
            else:
                return [name, '>', threshold]
    
    max_depth = model.get_params()['max_depth']
    bin_list = []
    for i in range(pow(2, max_depth)):
        bin_list.append(list(bin(i)[2:].rjust(max_depth,'0')))
    rule_all_list = []
    for j in range(pow(2, max_depth)):
        rule_list = []
        node = 0
        flag = bin_list[j][0]
        for k in range(max_depth):
            if k == 0:
                rule_list.append(recurse(node, k+1, flag))
            else:
                if flag == '0':
                    node = tree_.children_left[node]
                    if tree_.children_left[node] != -1:
                        flag = bin_list[j][k]
                        rule_list.append(recurse(node, k+1, flag))
                    else:
                        break
                else:
                    node = tree_.children_right[node]
                    if tree_.children_right[node] != -1:
                        flag = bin_list[j][k]
                        rule_list.append(recurse(node, k+1, flag))
                    else:
                        break
        rule_all_list.append(rule_list)

    unique_list = [rule_all_list[0]]
    for i in range(len(rule_all_list)-1):
        if rule_all_list[i] != rule_all_list[i+1]:
            unique_list.append(rule_all_list[i+1])
    
    return unique_list

#编写生成lift表函数，需要用到上个函数得到的规则列表
#lif表包括训练集和测试集的结果，将他们进行横向拼接得到df类型数据，并且按照训练集的fpd7进行升序排序
def df_lift(data, data1, rule_list): #输入参数：data是训练数据集，就是训练决策树的，data1是测试数据集，rule_list是规则列表
    f = lambda x: format(x, '.2%') #将小数字转化为百分比函数
    data_split_lift = pd.DataFrame([[data.shape[0], 1, data.is_fpd7.sum(), data.is_fpd7.mean(), 1]], index = ['all_data'], 
                                 columns = ['训练集总人数', '训练集人数占比', '训练集fpd7人数', '训练集fpd7', '训练集lift值'])
    data_split_lift1 = pd.DataFrame([[data1.shape[0], 1, data1.is_fpd7.sum(), data1.is_fpd7.mean(), data1.is_fpd7.mean()/data.is_fpd7.mean()]], index = ['all_data'], 
                             columns = ['测试集总人数', '测试集人数占比', '测试集fpd7人数', '测试集fpd7', '测试集lift值'])
    num = 1
    for i in rule_list:
        d =  data.copy()
        d1 =  data1.copy()
        for j in i:
            if j[1] == '<=':
                d = d[d[j[0]] <= j[2]]
                d1 = d1[d1[j[0]] <= j[2]]
            else:
                d = d[d[j[0]] > j[2]]
                d1 = d1[d1[j[0]] > j[2]]
        print(d)
        count = d.shape[0]
        count1 = d1.shape[0]
        count_per = count/data.shape[0]
        count_per1 = count1/data1.shape[0]
        fpd7_cnt = d.iloc[:, -1].sum()
        fpd7_cnt1 = d1.iloc[:, -1].sum()
        fpd7 = d.iloc[:, -1].mean()
        fpd71 = d1.iloc[:, -1].mean()
        lift = fpd7/data.iloc[:, -1].mean()
        lift1 = fpd71/data.iloc[:, -1].mean()
        d_split_lift = pd.DataFrame([[count, count_per, fpd7_cnt, fpd7, lift]], index = ['rule_'+str(num)], 
                                 columns = ['训练集总人数', '训练集人数占比', '训练集fpd7人数', '训练集fpd7', '训练集lift值'])
        d_split_lift1 = pd.DataFrame([[count1, count_per1, fpd7_cnt1, fpd71, lift1]], index = ['rule_'+str(num)], 
                                 columns = ['测试集总人数', '测试集人数占比', '测试集fpd7人数', '测试集fpd7', '测试集lift值'])
        data_split_lift = data_split_lift.append(d_split_lift)
        data_split_lift1 = data_split_lift1.append(d_split_lift1)
        num = num + 1

    data_lift = pd.concat([data_split_lift, data_split_lift1], axis = 1)
    data_lift_rule = data_lift.iloc[1:, :].sort_values(by = '训练集fpd7')
    data_lift_all = data_lift.iloc[:1, :]
    data_lift = pd.concat([data_lift_all, data_lift_rule], axis = 0)
    data_lift['测试集人数占比'] = data_lift['测试集人数占比'].apply(f)
    data_lift['训练集人数占比'] = data_lift['训练集人数占比'].apply(f)
    data_lift['测试集fpd7'] = data_lift['测试集fpd7'].apply(f)
    data_lift['训练集fpd7'] = data_lift['训练集fpd7'].apply(f)
    return data_lift

#此函数使用场景：想要将满足几条规则的样本筛选出来，对其进行进一步操作
#输入数据集和规则列表，以及需要使用的规则序号（如第1/3/5三条规则），自动输出满足目标规则的数据集
#注意：目标规则需要的输入方式，如[1,3,5]:代表第1/3/5三条规则
def df_rule(data, rule_list, rule_num_list):
    r_list = []
    for i in rule_num_list:
        r_list.append(rule_list[i-1])
    df = pd.DataFrame(columns = data.columns)
    for j in r_list:
        d = data.copy()
        for k in j:
            if k[1] == '<=':
                d = d[d[k[0]] <= k[2]]
            else:
                d = d[d[k[0]] > k[2]]
        df = df.append(d)
    return df

rule_list = tree_rule(model, x1.columns) #拆解决策树得到规则

# 分层效果校验——测试集效果检验
data = pd.concat([x1, y1], axis = 1) #训练集数据
data1 = pd.concat([x2, y2], axis = 1) #测试集数据
data_lift = df_lift(data, data1, rule_list) #获取测试集lift表
data_rule = df_rule(data, rule_list, [1]) #取出满足某些规则的数据集，一种方法而已，看自己需求使用

# 分层效果校验——逐月效果检验
# 获取所有授信年月信息，去重后升序排成列表
date_list = sorted(list(set(df.credit_submit_date_month.sort_values().tolist())))
month_list = []
for i in date_list:
    month_list.append(i[:7])
month_list = sorted(list(set(month_list)))

# 将训练集和测试集数据集合在一起，获得总数据，并将数据按照时间顺序按月切分数据，分别存储再data_month中（这里使用授信月作为时间维度：credit_submit_date_month，实际使用灵活选择）
# 轮流将每个月得数据作为测试集创建lift表，并依次存储在data_lift_month中
data_all = data.append(data1) #合并为总数据
data_month = [] #创建列表用来存储每个月的数据集
data_lift_month = [] # 创建列表存储每个月的lift检验表
# 下面是创建分月数据和相应lift表的过程（注：可以不以单个月份为单位，选择三个月/半年等均可，只需要在下面的代码中对循环的步长和判断时间大小时i的选择做更改即可）
for i in range(len(month_list)):
    data_month.append(data_all[(df.credit_submit_date_month >= month_list[i] + '-01') & (df.credit_submit_date_month <=  month_list[i] +'-31')]) #创建分月数据
    data_lift_month.append(df_lift(data, data_month[-1], rule_list)) #对分月数据创建Lift表，然后存储
```
